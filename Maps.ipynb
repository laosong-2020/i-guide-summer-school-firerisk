```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Path to the GeoJSON file
geojson_file = "C:\\Users\\harma\\OneDrive\\Desktop\\i-guide-summer-school-firerisk\\datasets\\results.geojson"

# Read the GeoJSON file
gdf = gpd.read_file(geojson_file)

# Print the GeoDataFrame
print(gdf)

# Plot the GeoDataFrame
gdf.plot()
```

```python
# Plot the GeoDataFrame
gdf.plot(figsize=(10, 8), column='WFIR_EALB', cmap='OrRd', legend=True)
plt.title('Visualization of WFIR_EALB Values')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
